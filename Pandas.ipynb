Python para análise de dados (Pandas)

```python
import pandas as pd
```

```python

df = pd.read_csv("/content/drive/MyDrive/datasets/Gapminder.csv", error_bad_lines=False, sep=";")
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
#Renomeando Colunas
df = df.rename(columns = {"country":"Pais", "continent": "Continente", "year": "Ano", "lifeExp": "Expectativa de vida", "pop": "Pop Total", "gdpPercap": "PIB"})
```

```python
#Total de linhas e colunas
df.shape
```

```python
#Visualizando as colunas
df.columns
```

```python
#Verificando tipo de dados
df.dtypes
```

```python
#Visualizando as últimas linhas
df.tail()
```

```python
#Informações e estatísticas sobre o conjunto de dados
df.describe()
```

```python
#Visualizando valores únicos + Filtro
df["Continente"].unique()
```

```python
#Visualizando valores únicos + Filtro
Oceania = df.loc[df["Continente"] == "Oceania"]
Oceania.head()
```

```python
#Visualizando valores únicos + Filtro
Oceania ["Continente"].unique()
```

```python
#Agrupamento de Dados
#Veficando a quantidade de país para cada continente

df.groupby("Continente")["Pais"].nunique()
```

```python
#Agrupamento de Dados
#Veficando a média de expectativa para cada ano
df.groupby("Ano")["Expectativa de vida"].mean()
```

```python
#Verificando a média de uma coluna
df["PIB"].mean()
```

```python
#Soma
df["PIB"].sum()
```
